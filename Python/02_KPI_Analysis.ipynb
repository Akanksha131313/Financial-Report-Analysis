### **KPI ANALYSIS-**

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/sample_data/financial_loan.xlsx")
```

```python
df
```

```python
# Total Loan Applications
total_apps = df['id'].count()
print("Total Loan Applications:", total_apps)
```

```python
# Total Funded Amount
total_funded = df['loan_amount'].sum()
print("Total Funded Amount:", total_funded)
```

```python
# Total Amount Received
total_received = df['total_payment'].sum()
print("Total Amount Received:", total_received)
```

```python
# Average Interest Rate
avg_int = df['int_rate'].mean() * 100
print("Average Interest Rate:", round(avg_int,2), "%")
```

```python
# Good Loan Metrics
good_loans = df[df['loan_status'].isin(['Fully Paid','Current'])]
print("\nGood Loan Count:", good_loans['id'].count())
```

```python
# Bad Loan Metrics
bad_loans = df[df['loan_status'] == 'Charged Off']

print("Bad Loan Count:", bad_loans['id'].count())
```
